```python
from sklearn.preprocessing import LabelEncoder 
le = LabelEncoder()
for i in train.columns:
    if train.dtypes[i] == "object":
        train[i] = le.fit_transform(train[i])
```

```python
train = train.drop(['IsActiveMember'], axis=1)
```

```python
from sklearn.model_selection import train_test_split
feature = train.iloc[:, 1:-2]
target1 = train.iloc[:,-2]
target2 = train.iloc[:,-1]
xtrain1, xtest1, ytrain1, ytest1 = train_test_split(feature, target1, test_size=0.2, random_state=0)
xtrain2, xtest2, ytrain2, ytest2 = train_test_split(feature, target2, test_size=0.2, random_state=0)
```
